### Read data from ADLS

####Order File

```python
df_ord = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("abfss://container1@stacecommerce.dfs.core.windows.net/Ecommerce_2000.csv.filepart")

```

```python
df_ord.display()
```

### Customer File

```python
df_cust = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("abfss://container1@stacecommerce.dfs.core.windows.net/dbo.customers.csv")

```

```python
df_cust.display()
```

###Write to Bronze Tables

```python
df_ord.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("e_commerce_catalog.bronze_schema.Orders_table")
```

```python
df_cust.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("e_commerce_catalog.bronze_schema.customers_table")
```
